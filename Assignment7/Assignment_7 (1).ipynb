```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
import os
```

```python
image_path = r"/content/WIN_20250710_14_14_08_Pro.jpg"
```

```python
image_path = r"/content/WIN_20250710_14_14_08_Pro.jpg"

def run_image_inference(model, image_path, save_path="output.jpg", colab=False):
    results = model(image_path)
    annotated_img = results[0].plot()
    results[0].save(save_path)
    if colab:
        plt.imshow(cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB))
        plt.axis("off")
        plt.show()
    else:
        cv2.imshow("YOLOv8 Detection - Image", annotated_img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

if __name__ == "__main__":
    model = YOLO("yolov8n.pt")
    run_image_inference(model, image_path, "result.jpg", colab=True)
```
